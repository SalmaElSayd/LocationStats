```python
class GoogleLocation:
      # init method or constructor
    def __init__(self, location_dic : dict):
        self.formatted_address = location_dic["formatted_address"]

    def __str__(self):
      return self.formatted_address

    def __repr__(self):
      return self.formatted_address

```

```python
#@title Default title text { run: "auto", vertical-output: true }
full_list = True #@param {type:"boolean"}
full_list
```

```python
color_dict = {chains.index[0] : 'green', chains.index[1] : 'red'}
chain_name = starbucks_kfc_uk_df["chain_name"]
colors = starbucks_kfc_uk_df.replace({"chain_name": color_dict})
colors=colors['chain_name']
locations = starbucks_kfc_uk_df[['latitude', 'longitude']]

if (full_list):
  sub =np.r_[0:len(starbucks_kfc_uk_df)]
else:
  sub= np.r_[0:10,718:721,900:910]

colors_sub = colors[sub]
locations_sub = locations.iloc[sub]
chain_name_sub = chain_name[sub]
locations_sub[10:12]
```
